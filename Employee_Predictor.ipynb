```python
from sklearn.preprocessing import LabelEncoder
```

```python
lr=LabelEncoder()
hr['salary']= lr.fit_transform(hr['salary'])
```

```python
Lor=LogisticRegression()
```

```python
ip=hr[['satisfaction_level','average_montly_hours','salary']]
```

```python
dp=hr['left']
```

```python
Lor.fit(ip,dp)
```

```python
ans = Lor.predict([[0.72,270,0]])
if ans[0]==1 :
  print('Employee can leave')
else:
  print('Employee cannot leave')
```

```python
from sklearn.metrics import accuracy_score
pval=Lor.predict(ip)
accuracy_score(dp,pval)
```
